# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
colors = ["#89CFF0","#FF69B4","#FFD700","#7B68EE","#FF4500","#9370DB","#32CD32",
          "#8A2BE2","#FF6347","#20B2AA","#FF69B4","#00CED1","#FF7F50","#7FFF00","#DA70D6"]
```

# Importing Data

```python
data = pd.read_excel("FOR ANALYSIS.xlsx")
data
```

```python
data.info()
```

```python
consumption = data[data['Avg. Cons / Month(Kg)'] >= 500]
consumption.head()
```

 # Six Months Average Consumption

```python
plt.figure(figsize=(15,6))
plt.bar(consumption["ITEM"],consumption["Avg. Cons / Month(Kg)"],color = colors,alpha=0.8)
plt.xticks(rotation=45)
plt.title('Items Avg Consumptions Last 6 Months',color='black',fontweight='bold',fontsize=14)
plt.xlabel('Items',color='black',fontweight='bold',fontsize=12)
plt.ylabel('Avg Consumption',color='black',fontweight='bold',fontsize=12)

for ITEM, count in zip(consumption['ITEM'],consumption['Avg. Cons / Month(Kg)']):
    plt.text(ITEM,count+5,str(count),ha='center',fontsize=8)
plt.show()
```

#### Graph shows that "Item 9" has maximum consumption during last 6 months i-e 143774 Kg.

# A,B,C ANALYSIS

```python
abc = data[['ITEM','TOTAL BUYING (Rs)']]
abc.head()
```

```python
abc['TOTAL BUYING (Rs)'].describe()
```

```python
category = []

for i in abc['TOTAL BUYING (Rs)']:
    if i >= 4.121400e+06:
        category.append("A")
    elif 1.064203e+06 <= i < 4.121400e+06:
        category.append('B')
    else:
        category.append('C')
print(category)        
```

```python
data['category'] = category
data.head()
```

```python
abc_items = data['category'].value_counts(ascending=True)
abc_items
```

```python
plt.figure(figsize=(14,8))
sns.set_style('dark')
ax = sns.countplot(data=data,x='category')
sns.despine(offset=10,trim=True)
plt.title('A,B,C Items Count',fontsize=16,fontweight='bold')
plt.xlabel("Category",fontsize=16,fontweight='bold')
plt.ylabel("Count",fontsize=16,fontweight='bold')
plt.xticks(fontsize=16,fontweight='bold')
for i in ax.containers:
    ax.bar_label(i,fontsize=16,fontweight='bold')
plt.show()    
```

#### Items that fall in A and B categories are 12.
#### 25 Items are fall in C category
